```python
# Use a GPU runtime
pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    revision="fp16",
    torch_dtype=torch.float16,
    use_auth_token=True
).to("cuda")
```

```python
 prompt = input("Write prompt to create a image: ")
with autocast("cuda"):
    image = pipe(prompt).images[0]

# Display
display(image)
```
